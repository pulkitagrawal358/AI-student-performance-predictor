```python
# 1.) Importing the neccessary files.
import pandas as pd
```

```python
# 2.) Loading the dataset.
df = pd.read_csv("/content/Student_performance_data _.csv")
```

```python
# 3.) Check for the missing values and removing the duplicates.
print("ðŸ”¹ Missing values before cleaning:")
print(df.isnull().sum())

duplicates = df.duplicated().sum()
print("\nðŸ”¹ Duplicate rows:", duplicates)

df = df.drop_duplicates()
```

```python
# 4.) Converting the categoricals to numericals.
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].astype('category').cat.codes

print("\nðŸ”¹ After converting categorical columns:")
print(df.head())
```

```python
# 5.) Feature & target selection.
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

print("\nðŸ”¹ Features shape:", X.shape)
print("ðŸ”¹ Target shape:", y.shape)
```

```python
# 6.) Saving the cleaned dataset.
df.to_csv("cleaned_student_data.csv", index=False)
print("\nâœ… Cleaned dataset saved")
```
